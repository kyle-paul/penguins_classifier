```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv("Dataset/penguins_size.csv")

df = df.dropna(subset=['body_mass_g'])
df.isnull().sum()

df['sex'].fillna('MALE', inplace=True)
df['sex'] = df['sex'].replace({'.': 'FEMALE'})
df.isnull().sum()
```

```python
df['sex'] = df['sex'].replace({'MALE': 0, 'FEMALE': 1})
df['island'] = df['island'].replace({'Torgersen': 0, 'Biscoe': 1, 'Dream': 2})
df['species'] = df['island'].replace({'Adelaie': 0, 'Chinstrap': 1, 'Gentoo': 2})

data = df.copy()
data = np.array(data)

X = data[:,1:]
y = data[:,0]

X.shape, y.shape
```

```python
from tensorflow.keras.utils import to_categorical

y_ohe = to_categorical(y, num_classes=3)
y.shape, y_ohe.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Input
from tensorflow.random import set_seed
from tensorflow.keras.backend import clear_session

clear_session()
set_seed(42)
np.random.seed(42)

model = Sequential()
model.add(Input(shape=X.shape[1:]))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')
model.summary()
```

```python
history = model.fit(X, y_ohe, epochs = 1500, verbose=1)
```

```python
import pickle
pickle.dump(model, open('model.pickle', 'wb'))
```
